```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("quatet.csv")
```

```python
cols = ['historical_sold', 'itemid', 'liked_count',
       'name', 'price', 'price_max', 'price_min', 'shop_location', 'shopid',
       'sold', 'stock', 'view_count']
df =df[cols]
df.sort_values(by = "price",ascending= False)
```

```python
df
```

```python
#plt.rcParams["figure.figsize"] = (18,3)
a = df.groupby(['shop_location']).sum()
```

```python

#plt.ylabel = "liked_count"
plt.xticks(rotation = 90)
plt.plot(a.index,a["historical_sold"])
plt.ylabel("sold Amount")
plt.xlabel('Location', fontsize=18)
plt.title("Item sell by location")
```

```python
a.sort_values(by="historical_sold")
```

```python
#plt.rcParams["figure.figsize"] = (18,3) # set the width and height of the plot
fig, ax = plt.subplots(figsize =(15,6))
plt.xticks(rotation=90)
ax.plot(a.index,a["sold"])
ax.set(title="Item sold by location",xlabel = "Location", ylabel="sold amount")
fig.savefig("sample.png")
```

```python
fig, ax = plt.subplots(figsize =(15,6))
plt.xticks(rotation=90)
ax.scatter(a.index,a["sold"])
ax.set(title="Item sold by location",xlabel = "Location", ylabel="sold amount")
#fig.savefig("sample.png")
```

```python
fig, ax = plt.subplots(figsize =(15,4))
plt.xticks(rotation=90)
ax.bar(a.index,a["sold"])
ax.set(title="Item sold by location",xlabel = "Location", ylabel="sold amount")
#fig.savefig("sample.png")
```
