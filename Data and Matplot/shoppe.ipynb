```python
import pandas as pd
import numpy as np
import requests
import time
```

```python
cols = ["itemid","shopid","name","image","stock","sold","historical_sold","liked_count","view_count","price","price_min","price_max","shop_location"]
final_df = pd.DataFrame(columns = cols)
```

```python
df = pd.DataFrame()
for i in range(0,2):
    url = f"https://shopee.vn/api/v4/search/search_items?by=relevancy&keyword=qu%C3%A0%20t%E1%BA%BFt&limit=60&newest={i*60}&order=desc&page_type=search&scenario=PAGE_GLOBAL_SEARCH&version=2"
    dicts = {}
    r = requests.get(url)
    data = r.json()
    time.sleep(15)
    for i in range(len(data["items"])):
        for item, value in data["items"][i]['item_basic'].items():
            for col in cols:
               if item == col:
                   dicts[col] = value

        df = df.append(dicts,ignore_index=True)
        
df["itemid"] = df["itemid"].astype(np.int64)
df["historical_sold"]= df["historical_sold"].astype(np.int64)
df["liked_count"]= df["liked_count"].astype(np.int64)
df["image"] = df.apply(lambda x:f"https://cf.shopee.vn/file/{x['image']}", axis=1)
df["price"] = df.apply(lambda x:x["price_min"]/100000, axis=1)
df["price"] = df["price"].astype(np.int64)
df["price_min"] = df.apply(lambda x:x["price_min"]/100000, axis=1)
df["price_min"] = df["price_min"].astype(np.int64)
df["price_max"] = df.apply(lambda x:x["price_max"]/100000, axis=1)
df["price_max"]= df["price_max"].astype(np.int64)
df["shopid"]= df["shopid"].astype(np.int64)
df["sold"]= df["sold"].astype(np.int64)
df["stock"]= df["stock"].astype(np.int64)
df["view_count"]= df["view_count"].astype(np.int64)

final_df = pd.concat([df],axis = 1)
#df.to_csv("data.csv",index=False)
```
