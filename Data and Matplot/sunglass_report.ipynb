```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sunglass.csv")
```

```python
df.head(5)
```

```python
df["brand_name"].unique()
```

```python
df_groupby = df.groupby(by=df["brand_name"]).sum()
df_groupby
```

```python
fig, ax = plt.subplots(figsize = (20,5));
plt.xticks(rotation = 90)
ax.scatter(df_groupby.index,df_groupby["final_price"]);
ax.plot(df_groupby.index,df_groupby["final_price"]);
ax.set(title="Price Comparision",xlabel = "Price", ylabel = "Brand_name");
```

```python
x = [3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot

```

```python
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=4, center=(5, 5),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 10), xticks=np.arange(1, 10),
       ylim=(0, 10), yticks=np.arange(1, 10));
```
