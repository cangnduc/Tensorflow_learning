```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart.csv")
df.head(5)
```

```python
over_50 = df[df["age"]>50]
over_50
```

```python
fig,ax =plt.subplots(figsize=(8,5))
scatter = ax.scatter(x= over_50["age"],y =over_50["chol"],c = over_50["oldpeak"])
ax.set(title="Over 50 Chol", xlabel = "Age",ylabel = "Chol")
ax.legend(*scatter.legend_elements(),title = "target");
```

```python
fig, (ax1,ax2) = plt.subplots(figsize=(10,10), nrows =2)
#scatter = ax1.scatter(x=over_50["age"],y= over_50["chol"],c= over_50["target"],label = "Heart")
scatter = ax1.bar(over_50["age"], over_50["chol"])
ax1.axhline(y=over_50["chol"].mean(),linestyle="--")
#ax1.legend(*scatter.legend_elements(),title="target")
ax1.set(title= "Heart Disease and Chol Level", xlabel = "Age", ylabel = "Chol Level");
scatter2 = ax2.scatter(over_50["age"], over_50["thalach"],c= over_50["oldpeak"])
ax2.axhline(over_50["thalach"].mean(),linestyle="--")
ax2.legend(*scatter2.legend_elements(),title="oldpeak");
ax2.set(title="Heart Disease and Thalach Level", xlabel= "age",ylabel="Thalach Level")
```
